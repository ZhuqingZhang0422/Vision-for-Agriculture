```python
import numpy as np
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 8) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# for auto-reloading external modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
from os import listdir
from os.path import isfile, join
mypath = 'jobs'
onlyfiles = [f for f in listdir(mypath) if (isfile(join(mypath, f)) and os.path.splitext(f)[-1].lower() == '.txt')]
```

```python
for f in onlyfiles:
    file = open(os.path.join(mypath,f))
    lines = file.readlines()
    res = []
    last = len(lines)-1
    for i in range (2,last):
        line = lines[i]
        epoch = float(line.split(',')[0].split(':')[1])
        train_loss = float(line.split(',')[1].split(':')[1])
        train_acc = float(line.split(',')[2].split(':')[1])
        train_IU = float(line.split(',')[3].split(':')[1])
        val_loss = float(line.split(',')[4].split(':')[1])
        val_acc = float(line.split(',')[5].split(':')[1])
        val_IU = float(line.split(',')[6].split(':')[1].split(' ')[1])
        res.append([epoch,train_loss,train_acc,train_IU,val_loss,val_acc,val_IU])
    data = np.array(res)
    df = pd.DataFrame(data,columns = ['Epoch','Train_loss','Train_acc','Train_IU','Val_loss','Val_acc','Val_IU'])
    csv_path = os.path.join(mypath,'Data_Process')
    df.to_csv(os.path.join(csv_path,f[:-4]+'.csv'))
    
    current_palette = sns.color_palette("Set2")
    # Figure 1
    plt.title('f[:-4]')
    plt.subplot(3,1,1)
    plt.plot(df['Epoch'],df['Train_loss'], '-o', color = current_palette[0])
    plt.plot(df['Epoch'],df['Val_loss'], '-o', color = current_palette[1])
    plt.ylabel('Loss', fontsize = 14)
    plt.legend(['Train_loss', 'Val_loss'], loc = 'upper right')
    
    # Figure 2
    plt.subplot(3,1,2)
    plt.plot(df['Epoch'],df['Train_acc'], '-^',color = current_palette[2])
    plt.plot(df['Epoch'],df['Val_acc'], '-^',color = current_palette[3])
    plt.ylabel('Acc', fontsize = 14)
    plt.legend(['Train_IU', 'Val_IU'], loc = 'upper right')
    
    # Figure 3
    plt.subplot(3,1,3)
    plt.plot(df['Epoch'],df['Train_IU'], '-*', color = current_palette[4])
    plt.plot(df['Epoch'],df['Val_IU'], '-*',color = current_palette[5])
    plt.xlabel('Epoch', fontsize = 14)
    plt.ylabel('Mean IU', fontsize = 14)
    plt.legend(['Train_IU', 'Val_IU'], loc = 'upper right')
    
    newpath = os.path.join(mypath,'Figure') 
    plt.savefig(os.path.join(newpath,f[:-4]+'.jpg'))
    plt.show()
        
```

```python
import re
mypath = '/Users/zhuqing/Desktop/Result'
models = [f for f in listdir(mypath) if (isfile(join(mypath, f)) and os.path.splitext(f)[-1].lower() == '.txt')]
```

```python
res, row = [] , []
for f in models:
    cur = []
    row.append(f[:-4])
    file = open(os.path.join(mypath,f))
    lines = file.readlines()
    for line in lines:
        num = re.findall(r"[-+]?\d*\.\d+|\d+", line)
        cur.append(float(num[0]))
    res.append(cur)
array = np.array(res)
print(row)
df = pd.DataFrame(array,index=row)
```

```python
#df.drop('Model',axis = 1, inplace = True)
df.astype(float)
df.columns = ['Mean_IOU','Background','Cloud_shadow','Double_plant','Planter_skip','Standing_water','Waterway','Weed_cluster']
```

```python
plt.rcParams['figure.figsize'] = (15, 8)
df.plot.bar(color = sns.color_palette("Set2"))
plt.title('IOU Score for different model',fontsize = 12)
plt.ylabel('IOU Score', fontsize = 12)
plt.xticks(rotation= 20, ha='right',fontsize = 12)
plt.savefig('Model_result.jpg')
```
